```python
import pandas as pd
import numpy as np
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Frank'],
    'Age': [25, np.nan, 30, 22, 35, np.nan],  # Two missing ages
    'Salary': [50000, 60000, 55000, np.nan, 70000, 62000], # One missing salary
    'Purchased': ['Yes', 'No', 'Yes', 'No', 'Yes', 'No'] # The Target Variable
}

df = pd.DataFrame(data)

print("--- Original Messy Data ---")
print(df)
```

```python

print(df.isnull().sum())
```

```python
mean_age = df['Age'].mean()
df['Age'] = df['Age'].fillna(mean_age)

print("--- After Filling Age ---")
print(df)
```

```python
df = df.dropna(subset=['Salary'])

print("--- Final Cleaned Data ---")
print(df)
```

```python
buyers = df[df['Purchased'] == 'Yes']
print(buyers)
```
